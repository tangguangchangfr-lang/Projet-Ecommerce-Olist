```python
#2. 读取 CSV 文件
# 2. read document CSV 
file_path = 'C:\\Users\\tangg\\OneDrive\\Desktop\\Data Fullstack Jedha\\Projet Supply Chain\\Data Olist\\data clean\\orders_clean.csv'
df_orders = pd.read_csv(file_path)
```

```python
#3. 导入到 MySQL
# if_exists='replace' 会删除旧表并创建新表
# if_exists='append' 会将数据追加到现有表中 (适合您的情况)
df_orders.to_sql(
    name='orders',          # 目标表名
    con=engine,             # 数据库连接
    if_exists='replace',     # 导入模式：追加
    index=False             # 不将 Pandas 索引导入为新列
)
```
